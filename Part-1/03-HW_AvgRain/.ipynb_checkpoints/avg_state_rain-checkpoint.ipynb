```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.colors as mcolors
```

```python
# Load in csv
filepath = 'Resources/avg_rain_state.csv'
data = pd.read_csv(filepath)
rain_df = pd.DataFrame(data)
rain_df.head()
```

```python
# Set x axis and tick locations
xaxis = np.arange(len(rain_df))
tick_locations = [value+0.45 for value in xaxis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(xaxis, rain_df["Inches"], color='skyblue', edgecolor='darkorange', alpha = 1, align='edge')
plt.xticks(tick_locations, rain_df['State'], rotation=55)
```

```python
# Set x and y limits
plt.xlim(-0.25, len(xaxis))
plt.ylim(0, max(rain_df['Inches'])+10)
```

```python
# Set a Title and labels
plt.title('Average Rainfall by State')
plt.xlabel('State')
plt.ylabel('Average Rainfall (inches)')
```

```python
# Save our graph and show the grap
plt.tight_layout
plt.show()
```
